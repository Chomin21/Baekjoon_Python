```python
#정수 N개의 합
def solve(a):
    sum = 0
    for e in a:
        sum = sum + e
    return sum
    
```

```python
#셀프 넘버
def d(n):
    term = 0
    if n<10 :
        term = n + n
    elif n>=10 & n<100:
        term = n + n//10 + n%10
    elif n>=100 & n<1000:
        term = n + n//100 + n%100//10 + n%100%10
    else:
        term = n + n//1000 + n%1000//100 + n%1000%100//10 + n%1000%100%10
    return term

bi = []
for i in range(1,10000):
    bi.append(d(i))
    #print(d(i))
    
last = []

for i in range(1,10000):
    t = 0
    while t < len(bi):
        if i == bi[t]:
            break;
        t = t+1
        if t == len(bi):
            last.append(i)
for i in last:
    print(i)
```

```python
#원준이 2번
def d(n):
    n = n + n//1000 + (n % 1000)//100 + (n % 100)//10 + (n % 10)
    return n

a = [0]*10000
for i in range(10000):
    a[i] = d(i)

for i in range(10000):
    if i in a:
        continue
    else:
        print(i)
```

```python
#한수
num = int(input())

def d(num):
    right = []
    k = 1
    while k <= num:
        if k<100 :
            right.append(k)
        else:
            if ((k//100) - ((k%100)//10)) == (((k%100)//10) - ((k%10))):
                right.append(k)
            else:
                pass
        k = k+1
    return len(right)

print(d(num))
```
